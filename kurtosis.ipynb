# import necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, kurtosis
import scipy.stats as stats
```

# Generate Uniform and gussian

```python
uniform = np.random.uniform(-1,1,1000)           
plt.figure(figsize=[20, 5]);                         
plt.title("uniform")                                      
plt.plot(uniform)                                           
plt.show()                                                 
```

```python
mean = 0                                                         
std = 1                                                        
num_samples = 1000                                             
wgn = np.random.normal(mean, std, size=num_samples)   

plt.figure(figsize=[25, 5]);                                   
plt.title("wgn")                                      
plt.plot(wgn)                                
plt.show()                                                      
```

# define Kurtosis function

```python
def kur(x):
    m = np.mean(x)
    m4 = (np.sum((x-m)**4))/len(x)
    m2 = (np.sum((x-m)**2))/len(x)
    k = (m4/(m2**2))
    return k-3
```

# my own kur function

```python
kur(wgn)
```

```python
kur(uniform)
```

# python kur function


```python
from scipy.stats import kurtosis
kurtosis(wgn) 
```

```python
kurtosis(uniform)
```

# difference function


```python
def diff(x,x_):
    return f"{x-x_:.25f}"
```

# calculate Error

```python
diff(kurtosis(wgn),kur(wgn))
```

```python
diff(kurtosis(uniform),kur(uniform))
```
